# Initial values 

| Parameter        | Name | Bounds |
| ---              | --- | --- |
| $\Delta\alpha_0$ | Difference between true and nominal position at $t_0$ | Normal(0,0.04) [mas] |
| $\Delta\delta_0$ | Difference between true and nominal position at $t_0$ | Normal(0,0.04) [mas] |
| $\mu_\alpha$     | Proper motion in RA direction                         | Constant value* [rad] |
| $\mu_\delta$     | Proper motion in Dec direction                        | Constant value* [rad] |
| $\varpi$         | Parallax                                              | Constant value* [rad] |
| $m_{planet}$     | Planet mass                                           | Constant value* [$M_\odot$] |
| $P_{orb}$        | Orbital period                                        | Uniform(0,5.5) [years] |
| $t_{peri}$       | Time of periastron passage                            | Uniform(0, $P_{orb}$) [years] |
| $e$              | Eccentricity                                          | Uniform(0, 0.5)    |
| $\Omega$         | Angle of ascending node                               | Uniform(0, 2$\pi$) [rad]|
| $\omega$         | Argument of periastron                                | Uniform(0, 2$\pi$) [rad]|
| $\cos i$         | Cosine of inclination                                 | Uniform(0,1) |

- *These values were calculated from N-body simulations using rebound. They were not changed to be accurate to Gaia or to any papers like the other variables were
- For $P_{orb}$ I was looking at periods from 0.1-10 but since this is focusing on visualising the different components of the astrometric signal I am restricting $P_{orb}$ to make sure we see a full orbit of the star around the barycenter 

```python
# no planet terms 

np.random.seed(8)

Delta_alpha_0 = np.random.normal(0, 1.93925472e-10)
Delta_delta_0 = np.random.normal(0, 1.93925472e-10)
mu_alpha = 2.3084641853871365e-07
mu_delta = 1.770935480191023e-07
parallax = 9.699321049402031e-08

# # planetary terms 
e = np.random.uniform(0, 0.5,1)
omega = np.random.uniform(0, 2*np.pi)
Omega = np.random.uniform(0, 2*np.pi)
cos_i = np.random.uniform(0, 1)

m_planet = 0.0143188
np.random.seed(8)
P_orb = 0.5#np.random.uniform(0, 5.5)
t_peri = np.random.uniform(0, P_orb)



# # other terms 
N = 70 #200
times = np.linspace(0, 5.5, N)
a_earth = 1
alpha0, delta0 = 1, 0.3
rad_mas = 206264806 # this factor changes units from radians to mas
```

```python
Delta_delta_0
```

# Finding terms 

In Ranalli 

$\Delta\eta(t)=\Delta\alpha\cos\theta(t)+\Delta\delta\sin\theta(t)$,

where 

$\Delta\alpha = \Delta\alpha_0 + \Pi_\alpha\varpi+(t-t_0)\mu_\alpha +(BX(t)+GY(t))$,

and 

$\Delta\delta = \Delta\delta_0 + \Pi_\delta\varpi+(t-t_0)\mu_\delta +(AX(t)+FY(t))$

# Proper Motion

```python
# Find proper motion
pm_term_ra, pm_term_dec = np.zeros((N)),np.zeros((N))

# for i in range(N):
pm_term_ra  = mu_alpha * times  + Delta_alpha_0
pm_term_dec = mu_delta * times  + Delta_delta_0
```

```python
# Checking functions 

from three_func_file_functions import generate_pm_signal

pm_term_ra_func, pm_term_dec_func = generate_pm_signal(Delta_alpha_0, Delta_delta_0, mu_alpha, mu_delta, times)

pm_sum_ra = np.sum(pm_term_ra - pm_term_ra_func*(1/4.8481368e12))
pm_sum_dec = np.sum(pm_term_dec - pm_term_dec_func*(1/4.8481368e12))


print("The difference between the file and the function in the RA direction is", pm_sum_ra)
print("The difference between the file and the function in the RA direction is", pm_sum_dec)
```

# Parallax

```python
# finding the parallax terms

d = a_earth/parallax
sind, cosd, sina, cosa  = sin(delta0), cos(delta0), sin(alpha0), cos(alpha0)

T = times

prlx_term_ra  = (a_earth/(d*cosd))*(sina*cos(2*np.pi*T)-cosa*sin(2*np.pi*T))               
prlx_term_dec = (a_earth/d)*sind*cos(2*np.pi*T-alpha0) 
```

```python
# Checking functions 
from three_func_file_functions import generate_parallax_signal

prlx_term_ra_func , prlx_term_dec_func = generate_parallax_signal(alpha0, delta0, parallax, times)

prlx_sum_ra = np.sum(prlx_term_ra - prlx_term_ra_func*(1/4.8481368e12) )
prlx_sum_dec = np.sum(prlx_term_dec - prlx_term_dec_func*(1/4.8481368e12) )

print("The difference between the file and the function in the RA direction is", prlx_sum_ra)
print("The difference between the file and the function in the RA direction is", prlx_sum_dec)
```

# Planet 

```python
# finding the planetary terms 
G_const=4*np.pi**2
m_star=1

# find_var_from_param
var1 = np.sqrt(cos_i)*cos(Omega) # [unitless]
var2 = np.sqrt(cos_i)*sin(Omega) # [unitless]
var3 = np.sqrt(e)*cos(omega)     # [unitless]
var4 = np.sqrt(e)*sin(omega)     # [unitless]


n = 2*np.pi/P_orb # [1/years]

a_AU = (G_const*(m_star+m_planet)/n**2)**(1/3) # [AU]

a_hat = (m_planet/(m_star+m_planet))*a_AU*parallax # [?]

sin_i = np.sqrt(1-cos_i**2) # [unitless]

A = a_hat * (  cos(omega)  * cos(Omega) - sin(omega) * sin(Omega) * cos_i)  # [?]
F = a_hat * ( - sin(omega) * cos(Omega) - cos(omega) * sin(Omega) * cos_i) # [?]

B = a_hat * (  cos(omega)  * sin(Omega) + sin(omega) * cos(Omega) * cos_i)   # [?]
G = a_hat * ( - sin(omega) * sin(Omega) + cos(omega) * cos(Omega) * cos_i) # [?]

C = a_hat * sin(omega)*sin_i
H = a_hat * sin_i*cos(omega)
    
    
sind, cosd, sina, cosa  = sin(delta0), cos(delta0), sin(alpha0), cos(alpha0)


B, A, F, G, H, C, _ = calculate_thiele_innes(omega, Omega, cos_i, parallax, m_planet, P_orb)

M = (2*np.pi)*(times - t_peri)/P_orb  

E = np.vectorize(rebound.M_to_E)(e,M)

X = (cos(E)-e)
Y = np.sqrt((1-e**2))*sin(E)

DELTA_X = A*X+F*Y            
DELTA_Y = B*X+G*Y            
DELTA_Z = H*X+C*Y  

plnt_term_ra  = (1/cosd) * (sina*DELTA_X-cosa*DELTA_Y)
plnt_term_dec = (-cosd*DELTA_Z + sind*(DELTA_X*cosa+DELTA_Y*sina))
```

```python
var4
```

```python
# Checking functions 
from three_func_file_functions import generate_planet_signal

plnt_term_ra_func , plnt_term_dec_func = generate_planet_signal(alpha0, delta0, parallax, var1, var2, var3, var4, m_planet, P_orb, t_peri, times)

plnt_sum_ra = np.sum(plnt_term_ra - plnt_term_ra_func*(1/4.8481368e12))
plnt_sum_dec = np.sum(plnt_term_dec - plnt_term_dec_func*(1/4.8481368e12))

print("The difference between the file and the function in the RA direction is", plnt_sum_ra)
print("The difference between the file and the function in the RA direction is", plnt_sum_dec)
```

# Plotting 

```python
# Create a 2x2 grid of plots
scale = 1.5
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10*scale, 8*scale))
axes = [ax1, ax2, ax3, ax4]

# Proper motion plot
ax1.plot(prlx_term_ra*rad_mas, 
         prlx_term_dec*rad_mas, 
         color='dodgerblue', 
         lw=1.5,           
         ls='--', 
         label = 'Working File')


# Planetary plot
ax2.plot(plnt_term_ra*rad_mas, 
         plnt_term_dec*rad_mas, 
         color='dodgerblue', 
         lw=1.5,           
         ls='--', 
         label = 'Working File')


# Parallax plot
ax3.plot(prlx_term_ra_func, 
         prlx_term_dec_func, 
         color='black', 
         lw=1.5,           
         ls='--', 
         label = 'Function')

# Planetary plot
ax4.plot(plnt_term_ra_func, 
         plnt_term_dec_func, 
         color='black', 
         lw=1.5,           
         ls='--', 
         label = 'Function')


axes = [ax1, ax2, ax3, ax4]
titles = ['Parallax', 'Planet', 'Parallax Function', 'Planet Function']
small_font = 15
big_font = 20

# Create and customize legends for all subplots in a loop
for i, ax in enumerate(axes):
    legend = ax.legend(loc="best", frameon=True, markerscale=1, fontsize=small_font)
    legend.get_frame().set_edgecolor('0.3')
    legend.get_frame().set_linewidth(1)
    
    # Set titles for each subplot
    ax.set_title(titles[i], fontsize= big_font)
    ax.set_xlabel('Right ascension (mas)', fontsize= small_font)
    ax.set_ylabel('Declination (mas)', fontsize= small_font)

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# Create a 2x2 grid of plots
scale = 1.5
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10*scale, 8*scale))
axes = [ax1, ax2, ax3, ax4]

# Proper motion plot
ax1.plot(pm_term_ra*rad_mas, 
         pm_term_dec*rad_mas, 
         color='black', 
         lw=1.5,           
         ls='--', 
         label = 'Model')


# Parallax plot
ax2.plot(prlx_term_ra*rad_mas, 
         prlx_term_dec*rad_mas, 
         color='black', 
         lw=1.5,           
         ls='--', 
         label = 'Model')


# Planetary plot
ax3.plot(plnt_term_ra*rad_mas, 
         plnt_term_dec*rad_mas, 
         color='black', 
         lw=1.5,           
         ls='--', 
         label = 'Model')


# Full signal plot
signal_ra = pm_term_ra + prlx_term_ra + plnt_term_ra
signal_dec = pm_term_dec + prlx_term_dec + plnt_term_dec

ax4.plot(signal_ra*rad_mas, 
         signal_dec*rad_mas, 
         color='black', 
         lw=1.5,           
         ls='--', 
         label = 'Model')


axes = [ax1, ax2, ax3, ax4]
titles = ['Proper Motion', 'Parallax', 'Planetary', 'Full signal']
small_font = 15
big_font = 20

# Create and customize legends for all subplots in a loop
for i, ax in enumerate(axes):
    legend = ax.legend(loc="best", frameon=True, markerscale=1, fontsize=small_font)
    legend.get_frame().set_edgecolor('0.3')
    legend.get_frame().set_linewidth(1)
    
    # Set titles for each subplot
    ax.set_title(titles[i], fontsize= big_font)
    ax.set_xlabel('Right ascension ($\mu$as)', fontsize= small_font)
    ax.set_ylabel('Declination ($\mu$as)', fontsize= small_font)

# Display the plots
plt.tight_layout()
plt.show()
```
